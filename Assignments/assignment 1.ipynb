## Generate 5000 two-dimensional  samples from a normal distribution with  mean $x=[10,20]$ and covariance matrix 
$S=\begin{bmatrix}
    3& 4  \\
    4& 8 \\
\end{bmatrix}$

## Perform the following steps

- generate the scatter plot of these two-dimensional data
- center the data 
- generate the scatter plot of after centering the data
- compute the sample covariance matrix and compare it with the acual covariance of the distribution
- are these two covariance matrices identical? if not what is the best way to make them as close of as possible ?
- compute the eigen vectors and values of the covariance matrix using np.linalg.eig; this way you decompose the covariance martix, i.e.
$S = E DE^{-1}$ where $E$ is the eigen vector matrix and $D$ is a diagonal matrix whose elements are eigen values. If you have no knowledge of eigen vectors and values, don't worry. The main objective of the assignment is not about eigen vectors and values. We will talk about them later in the course.
- transform the data matrix by $E^T$, i.e. , $X_t = E^T X$

- generate the scatter plot of the two-dimensional data $X_t$. What change you observe in the scatter plot and what is the implication of this transformation?
- transform the data matrix $X_t$ by $D^{-0.5}$, i.e. , $X_w = D^{-0.5} X_t$
- generate the scatter plot of the two-dimensional data $X_w$ .What change you observe in the scatter plot and what is implication of this transformation?

# provide the code, explanation, and scatter plot in an ipynb file such that we can run it on Jupyter notebook. 


```python
import numpy as np
import matplotlib.pyplot as plt

mean = [10, 20]
covariance = np.matrix([[3, 4], [4, 8]])
X = np.random.multivariate_normal(mean, covariance, 5000)
plt.scatter(X[:, 0], X[:, 1], c='red', edgecolors='black')
plt.show()
X = X - mean
plt.scatter(X[:, 0], X[:, 1], c='yellow', edgecolors='black')
plt.show()
sample_covariance = np.cov(X[:,0], X[:, 1])
print('Covariance: \n', covariance)
print('Sample Covariance: \n', sample_covariance)

"""
The covariance and sample covariance are not exactly identical, but they are extremely close to each other.
In order to make them as close as possible to each other, the best way to do that would be to increase
the amount of samples that are being generated. 5000 is a lot, so they are pretty close. However, if you were
to instead do only 5 for example, they tend to not be as close to each other. 
"""

eigen, D = np.linalg.eig(covariance)
XT = X * eigen.T
plt.scatter(XT[:, 0], XT[:, 1], c='green', edgecolors='black')
plt.show()

"""
Transforming the matrix in this way led to the y values being much more spread out. The scale in the y-axis is
further zoomed out on the scatter plot to compensate for this, but the x-axis values essentially stayed almost
the same. They were multiplied by a number that was close enough to one that the difference isn't noticeable
relative to the change in y-axis values. This transformation made it so the variance in the y-axis values have
a higher variance. The y-values on the scatter plot in this case are the 2nd dimmension of vector Y. (Second column).
"""

D = np.absolute(D)
D = 1/np.sqrt(D)
XW = XT * D
plt.scatter([XW[:, 0]], [XW[:, 1]], c='pink', edgecolors='black')
plt.show()

"""
As a result of this transformation, the scatter plot becomes nearly perfectly linear, meaning the data is now
more directly correlated to each other. 
"""
```
