```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

mat = scipy.io.loadmat('yeastdata.mat')
genes = mat['genes']
times = mat['times']
yeastvalues = mat['yeastvalues']
```

```python
times = times.reshape(7)
```

```python
plt.plot(times, yeastvalues[15, :])
plt.xlabel('Time (Hours)')
plt.ylabel('Log2 Relative Expression Level')
plt.ylim(-1, 5)
plt.xlim(0, 25)
plt
```

```python
plt.plot(times, (2**yeastvalues[15, :]))
plt.xlabel('Time (Hours)')
plt.ylabel('Log2 Relative Expression Level')
plt.ylim(-1, 5)
plt.show()
```

```python
# 15 so that it shows the above graph as well.

for i in range(15, 26):
    plt.plot(times, (2**yeastvalues[i, :]))
    plt.xlabel('Time (Hours)')
    plt.ylabel('Log2 Relative Expression Level')
    plt.title('Profile Expression Levels')
    plt.xlim(0, 25)
    plt.ylim(0, 20)
plt.show()
```
